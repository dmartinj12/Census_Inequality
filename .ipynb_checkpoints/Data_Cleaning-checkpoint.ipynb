```python
import pandas as pd
```

```python
# Read in census data
df = pd.read_csv("Income_by_Race_Region.csv")
# Filter out columns that contain 'Margin of Error' in their names
df_filtered = df.loc[:, ~df.columns.str.contains("!!Margin of Error")]
df_filtered = df_filtered.loc[:, ~df_filtered.columns.str.contains("Percent Distribution")]
df_filtered = df_filtered.loc[:, ~df_filtered.columns.str.contains("Number!!Estimate")]
# Save or inspect the filtered DataFrame
df_filtered.to_csv("filtered_dataset.csv", index=False)
df_filtered.head()
```

```python
def extract_state(column_name):
    # Extract the state name from columns
    return column_name.split("!!")[0] if "!!" in column_name else column_name

# Rename columns by applying the extract_state function
df_filtered.columns = [extract_state(col) for col in df_filtered.columns]
df_filtered.head(35)
```

```python
# Drop irrelevant columns
df_filtered = df_filtered.drop(index=range(21)).reset_index(drop=True)
df_filtered = df_filtered.drop([1], axis=0)
df_filtered.head(10)
```

```python
#Export to CSV file
df_filtered.to_csv("Race_n_Region_cleaned.csv", index=False)
```
